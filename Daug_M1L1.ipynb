#### 1.1 More About Programs 

- A program is a series of instructions that determines how computation can be performed. The measurement may be anything as complex as making a web browser Html page or encoding a video and streaming it across the network. It may also be a symbolic calculation, such as looking for and replacing text in a document or compiling a program (strangely enough).
- In different languages, the specifics look different, but a few simple instructions appear in just about every language.
    - Input 
        - Get data from a keyboard, a file, or from some other computer.
    - Output
        - View on-screen data or transfer data to a file or other computer.
    - Math and logic
        - Perform basic mathematical activities such as adding and multiplying, and logical operations such as and, or, and not.
    - Conditional execution
        - Check for certain conditions and perform the appropriate statement sequence.
    - Repetition
        - Perform some action repeatedly, usually with some variation.

Thus, once the subtasks are easy enough to be done with sequences of these basic instructions, we can define programming as the process of breaking a large, complex task into smaller and smaller subtasks.

#### 1.2 Review of Control Structures 

We have structures first, enabling us to iterate over something. We may look at character-by-character strings or item-by-item lists until we have reached the end of them by using something called a loop for.

Example:

```python
for character in "Cool string":
 print(character)
```

- With while loops, we can also iterate without a definite stop point.

Example:

```python
grocery_item = ""
while grocery_item != "done":
 grocery_item = input("Please write down an item to add to your grocery list. When you are done writing the list simply type: done")
 print(grocery_item)
```

- Other systems would only allow us to run parts of our programs or perform certain tasks if a certain set of conditions is identified. Conditionals allow us to do that, as they're called.

Example:

```python
grocery_item = ""
grocery_list = []
while grocery_item != "done":
    grocery_item = input("Please write down an item to add to your grocery list. When you are done writing the list then simply type: done")
    if grocery_item == 'done':
        continue
    else:
        print("adding the item to the list")
        grocery_list.append(grocery_item)
print("Here is our grocery list:")
print(grocery_list)
```

#### 1.3 Formal and Natural Languages

- Natural Languages - These are the languages spoken by people, such as Korean, English, Spanish, and Mandarin Chinese. Individuals did not create them (although people try to impose some order on them); they evolved naturally.
- Formal Languages - languages that are designed by people for specicfic application.
- 2 Syntax Rules:
    -Tokens - basic elements in language.
    -Structure - the way the tokens are arranged.
    
- Parsing - the process of figuring out the structure.

- Semantics - the meaning of the structure

- Differences between Formal and Natural languages:

    - ambiguity
        - natural languages are full of ambiguity.
        - formal languages are designed to be nearly or completely unambiguous, which mean that any statement has exactly one meaning, regrdless of context.
    - redundancy
        - natural languages employ lots of redundancy.
        - formal are less redundance and more concise.
    - literalness
        - natural are full of idiom and methaphor.
        - formal means exactly what they say.
    - poetry
        - Words are used both for their sounds and for their significance, and an effect or emotional response is generated by the whole poem together. Ambiguity is not only widespread but also intentional.
    - prose
        - The literal sense of words is more important, and the structure gives more meaning. The prose is more open to analysis than poetry, but it is also always vague.
    - program
        - A computer program's meaning is unambiguous and literal and can be completely understood by studying the tokens and structure.
Note: To read programs remember that formal languages are much denser, structure is important (learn the process parse, tokens, structure), and details matter.
