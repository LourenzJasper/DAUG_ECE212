## Abstraction

### 1 Values and Data Types

- A value is one of the fundamental things — like a word or a number — that a program manipulates. The values we have seen so far are 5 (the result when we added 2 + 3), and "Hello, World!". We often refer to these values as objects and we will use the words value and object interchangeably.
- These objects are classified into different classes, or data types: 4 is an integer, and "Hello, World!" is a string, so-called because it contains a string or sequence of letters. You (and the interpreter) can identify strings because they are enclosed in quotation marks.
- If you are not sure what class a value falls into, Python has a function called type which can tell you.
- Numbers with a decimal point belong to a class called float, because these numbers are represented in a format called floating-point. At this stage, you can treat the words class and type interchangeably. We’ll come back to a deeper understanding of what a class is in later chapters.

### 1.1 Opeartors and Operands
- Operators
    - can build complex expressions out of simpler ones. It represents computations like addition, multiplication, abd division.
    
- Operands
    - values that operator works
    
- Legal Python Expressions
    - (+)addition
    - (-) subtraction
    - (/) division
    - (*) multiplication
    - (**) exponentiation
    - (//) truncated division
    - (%) modulus operator/remainder operator

### 1.2 Function Calls
- The Python interpreter can compute new values with function calls. Functions are like factories that take in some material, do some operations, and then send out the resulting object.
- The materials as arguments or inputs and the resulting object as output or return value.

Note: The output of a function is a Python value and we can never really  see the internal representation of a value. 

### 1.2.1 Function calls as part of complex expressions
- Wherever you can write a literal as a number in an expression, you can also write a function invocation that produces a number.


### 1.2.2 Functions are objects: parenthesis invoke functions
- Functions are just artifacts themselves. If you tell Python to print the function object, you'll get one of those not-so-nice printed representations instead of printing the results of invoking the function object.

- Only typing the function's name refers to the function as an object. The function is invoked by typing the function name followed by the parenthesis ().

The	surface	of	the	Earth	is	curved,	and	the	distance	between	degrees	of	longitude
varies	 with	 latitude.	 As	 a	 result,	 finding	 the	 distance	 between	 two	 points	 on	 the	
surface	of	the	Earth	is	more	complicated	than	simply	using	the	Pythagorean	theorem.
Let	(t1,	g1)	and	(t2,	g2)	be	the	latitude	and	longitude	of	two	points	on	the	Earth’s
surface.	The	distance	between	these	points,	following	the	surface	of	the	Earth,	in
kilometers	is:

distance	 = 6371.01 arccos(sin(t1) × sin(t2) + cos(t1) × cos(t2) × cos(g1 − g2))

The	value	6371.01	in	the	previous	equation	wasn’t	selected	at	random.	It	is	the
average	radius	of	the	Earth	in	kilometers
